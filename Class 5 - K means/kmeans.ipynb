```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets  import make_blobs
```

```python
X,y = make_blobs(n_samples=900 , n_features=2 , centers=5 ,  random_state=3)
```

```python
k = 5

color  = ["green" , "red" , "blue" , "yellow" , "orange"]

clusters = {}

np.random.seed(3)
for i in range(k):
    center = 10*(2*np.random.random((2 , ))-1)
    point = []
    cluster = { 
        'center' : center , 
        'points' : point ,
        'color'  : color[i]
    }
    
    clusters[i] = cluster
    
    
print(clusters)
```

```python
def distance(x1 , x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
def assignPointToCluster(clusters):
    for i in range(X.shape[0]):
        dist  = []
        curr = X[i]
        
        for kx in range(k):
            d = distance(curr , clusters[kx]["center"])
            dist.append(d)
        current_cluster = np.argmin(dist)
        #print(curr)
        clusters[current_cluster]["points"].append(curr)
```

```python
def updateCluster(clusters):
    for kx in range(k):
        pts = np.array(clusters[kx]["points"])
        
        if pts.shape[0]>0:
            new_u = pts.mean(axis = 0)
            clusters[kx]["center"] = new_u
            clusters[kx]["points"] = []
```

```python
def plotcluster(clusters):
    for kx in range(k):
        pts = np.array(clusters[kx]["points"])
        try:
            plt.scatter(pts[:,0] , pts[:,1] , color = clusters[kx]["color"])
        except:
            pass
        
        
        uk = clusters[kx]["center"]
        #print(uk)
        plt.scatter(uk[0] , uk[1] , color = "black",  marker = "*")
```

```python
assignPointToCluster(clusters)
plotcluster(clusters)
```

```python
for i in range(100):
# Repeat this cell few times or put it inside a loop
    assignPointToCluster(clusters)
    plotcluster(clusters)
    updateCluster(clusters)
```
