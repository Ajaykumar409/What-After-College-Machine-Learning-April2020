```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = np.arange(10)
Y = (X-5)**2
```

```python
X ,Y
```

```python
plt.plot(X,Y)
```

```python
lr = 0.1
x  = 8
error = []
plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5)
    x = x - lr*grad
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)
    print(x)
plt.show()
```
